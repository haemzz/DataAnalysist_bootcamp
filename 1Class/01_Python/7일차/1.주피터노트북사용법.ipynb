# <font color = blue>마크다운
마크다운은 HTML 기반이므로, 태그를 이용하여 글자 색깔을 바꾸거나, 수식을 넣거나, 이미지를 넣는 등의 방법으로도 활용할 수 있다.<br>
- 글자 색깔 바꾸는 방법: \<font color = blue> # 원하는 색깔로
- 이미지 넣는 방법: 이미지를 복사 후 마크다운 상태에서 붙여넣기를 하면 해당 명령어가 입력된다.
- 수식 넣기: 넣을 수식의 양옆에 \$ (달러표시)를 해준다. 분수형태를 만들고 싶을 때에는 \'\cfrac{}'을 사용한다.  
    \<sup>: 윗첨자 \<sub>: 아래첨자

# <font color = blue>주피터 노트북 단축키
* shift + enter: 액티브 셀을 실행하고 ^아래 셀을 선택^
* ctrl + enter: 액티브 셀을 실행
* alt + enter: 액티브 셀을 실행하고 아래에 셀을 하나 생성
* shift + ctrl + - : ^커서 위치에서^ 셀을 둘로 나눔
* shift + m: 선택 셀과 아래 셀을 합침
* dd: 선택 셀 삭제
* a: 위로 셀 추가
* b: 아래로 셀 추가
* m: markdown으로 변경
* y: code로 변경
